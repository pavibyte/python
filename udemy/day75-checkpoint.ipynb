THE MONTH COLUMN IN THE TESLA DATA SET IS OF TYPE str SO NOW WE NEED TO CONVERT IT INTO DATETIME BY THE FUNTION .to_datetime()

```python
df_tesla.MONTH = pd.to_datetime(df_tesla.MONTH)
```

```python
df_tesla.MONTH.head
```

.gca() get current axis to have a common x axis and two y axis on both left and right to plot different data.

```python
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.set_ylabel('TSLA stock price',color='g')
ax2.set_ylabel('TSLA search trend')

ax1.plot(df_tesla.MONTH,df_tesla.TSLA_USD_CLOSE,color='g')
ax2.plot(df_tesla.MONTH,df_tesla.TSLA_WEB_SEARCH)
```

```python
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.set_ylabel('TSLA stock price',color='g')
ax2.set_ylabel('TSLA search trend')

ax1.plot(df_tesla.MONTH,df_tesla.TSLA_USD_CLOSE,color='g')
ax2.plot(df_tesla.MONTH,df_tesla.TSLA_WEB_SEARCH,linestyle='dotted')
```
